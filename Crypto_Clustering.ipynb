```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
# Instantiate the StandardScaler
scaler = StandardScaler()

# Fit and transform the data
df_market_data_scaled = pd.DataFrame(scaler.fit_transform(df_market_data), columns=df_market_data.columns)

# Display the scaled data
df_market_data_scaled.head(10)
```

```python
# Create a DataFrame with the scaled data
df_scaled = pd.DataFrame(scaler.fit_transform(df_market_data), columns=df_market_data.columns)

# Copy the crypto names from the original data
df_scaled['crypto_names'] = df_scaled.index

# Display sample data
df_scaled.head(10)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))

# Display the list
print(k_values)
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    # Create a KMeans model with explicit n_init
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    
    # Fit the model to the scaled data
    kmeans.fit(df_scaled)
    
    # Append the model.inertia_ to the inertia list
    inertia_values.append(kmeans.inertia_)

# Display the inertia values
inertia_values
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_dict = {'k_values': k_values, 'inertia_values': inertia_values}

# Create a DataFrame with the data to plot the Elbow curve
elbow_data_df = pd.DataFrame(elbow_data_dict)

# Display the DataFrame
elbow_data_df
```

```python
# Import necessary libraries for plotting
import hvplot.pandas

# Create a DataFrame from the dictionary
elbow_data_df = pd.DataFrame(elbow_data_df)

# Plot a line chart
elbow_data_df.hvplot.line(x='k_values', y='inertia_values', title='Elbow Curve', xlabel='Number of Clusters (k)', ylabel='Inertia', line_width=3)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 2

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
best_k = 2
```

```python
# Initialize the K-Means model with the best value for k
best_kmeans_model = KMeans(n_clusters=best_k, n_init=10, random_state=42)
```

```python
# Fit the model to the scaled data
best_kmeans_model.fit(df_scaled)

# Get the cluster labels for each data point
cluster_labels = best_kmeans_model.labels_

# Display the cluster labels (optional)
print(cluster_labels)
```

```python
# Create a copy of the DataFrame
df_copy = df_scaled.copy()

# Display sample data from the copied DataFrame
df_copy.head()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_copy['predicted_clusters'] = best_kmeans_model.predict(df_scaled)

# Display sample data with the new column
df_copy.head()
```

```python
# Import necessary libraries
import hvplot.pandas

# Set the features for x and y axes
x_feature = "price_change_percentage_24h"
y_feature = "price_change_percentage_7d"

# Create the scatter plot with K-Means labels and crypto names
scatter_plot = df_copy.hvplot.scatter(
    x=x_feature,
    y=y_feature,
    c='predicted_clusters',  
    colormap='viridis',      
    hover_cols=['crypto_names'],  
    title='Scatter Plot with K-Means Clusters',
    xlabel='Price Change Percentage (24h)',
    ylabel='Price Change Percentage (7d)'
)

# Show the plot
scatter_plot
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance with n_components=3
pca_model = PCA(n_components=3)

# Fit the PCA model to the scaled data
pca_result = pca_model.fit_transform(df_scaled)

# Display the explained variance ratio for each principal component
print("Explained Variance Ratio:", pca_model.explained_variance_ratio_)

```

```python
# Use the PCA model with fit_transform to reduce to three principal components
pca_result = pca_model.fit_transform(df_scaled)

# Create a DataFrame with the reduced data
df_pca = pd.DataFrame(data=pca_result, columns=['Principal_Component_1', 'Principal_Component_2', 'Principal_Component_3'], index=df_scaled.index)

# View the first five rows of the DataFrame
df_pca.head()
```

```python
# Retrieve the explained variance ratio for each principal component
explained_variance_ratio = pca_model.explained_variance_ratio_

# Display the explained variance ratio
print("Explained Variance Ratio:")
for i, ratio in enumerate(explained_variance_ratio, start=1):
    print(f"Principal Component {i}: {ratio:.4f}")
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 0.9881

```python
# Creating a DataFrame with the PCA data
df_pca_final = pd.DataFrame(data=pca_result, columns=['Principal_Component_1', 'Principal_Component_2', 'Principal_Component_3'], index=df_scaled.index)

# Copy the crypto names from the original data
df_pca_final['crypto_names'] = df_scaled.index

# Set the 'coin_id' column as the index
if 'coin_id' in df_market_data.columns:
    df_pca_final.set_index('coin_id', inplace=True)

# Display sample data from the new DataFrame
df_pca_final.head()
```

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a DataFrame with the PCA data 
df_market_data_pca = pd.DataFrame(data=pca_result, columns=['Principal_Component_1', 'Principal_Component_2', 'Principal_Component_3'], index=df_scaled.index)

# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    # Create a KMeans model with explicit n_init
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    
    # Fit the model to the PCA data
    kmeans.fit(df_market_data_pca)
    
    # Append the model.inertia_ to the inertia list
    inertia_values.append(kmeans.inertia_)

# Display the inertia values
print(inertia_values)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {'k_values': k_values, 'inertia_values': inertia_values}

# Display the dictionary
elbow_data

# Create a DataFrame with the data to plot the Elbow curve
elbow_data_df = pd.DataFrame(elbow_data)

# Display the DataFrame
elbow_data_df

```

```python
# Import necessary libraries for plotting
import hvplot.pandas

# Create a DataFrame with the data to plot the Elbow curve
elbow_data_df = pd.DataFrame({'k_values': k_values, 'inertia_values': inertia_values})

# Plot a line chart
elbow_data_df.hvplot.line(x='k_values', y='inertia_values', title='Elbow Curve', xlabel='Number of Clusters (k)', ylabel='Inertia', line_width=3)
```

### Cluster Cryptocurrencies with K-means Using the PCA Data
